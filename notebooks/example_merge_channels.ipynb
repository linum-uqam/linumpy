# Exemple d'utilisation de merge channel
Principalement utilis√© pour visualiser des annotations

```python
import tifffile as tf
import matplotlib.pyplot as plt

"""
# Might need to import linumpy location in module path if not done already
import sys
sys.path.insert(0, "/path/to/linumpy-repository")
"""

# local module
from linumpy import viz
```

```python
anno_path = 'C:/Users/User/Desktop/Etude/LINUM/Carreno segmentation/annotation/'
cell_path = 'C:/Users/User/Desktop/Etude/LINUM/Carreno segmentation/crop/'

s1_anno = tf.imread(anno_path + 'slik1.tif')
s1_crop = tf.imread(cell_path + 'Slik GFP #01.tif')
```

```python
z = 11
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(s1_crop[z])
plt.title('crop')
plt.subplot(122)
plt.imshow(s1_anno[z])
plt.title('anno')
plt.show()
```

```python
plt.figure(figsize=(15,15))

plt.subplot(131)
plt.imshow(viz.merge_channel(s1_crop, red=s1_anno, ratio=0.70)[z])
plt.title('red')

plt.subplot(132)
plt.imshow(viz.merge_channel(s1_crop, green=s1_anno, ratio=0.70)[z])
plt.title('green')

plt.subplot(133)
plt.imshow(viz.merge_channel(s1_crop, red=s1_anno, green=s1_anno, ratio=0.70)[z])
plt.title('yellow')

plt.show()
```
