```python
import matplotlib.pyplot as plt
import numpy as np
import nibabel as nib
```

## Extraction de la ROI

```python
#Importation
from linumpy import util
```

```python
#Nom des fichiers
dir_path = "../../Data/ROI_along_AC/"
ROI_name = "ROI_01.nii"
brain_name = "mice01-2xy.nii"
```

```python
#Ouvrir fichiers
brain = nib.load(dir_path + brain_name)
ROI_mask = nib.load(dir_path + ROI_name)

# Extraction et sauvegarde de la ROI
# Le fichier extracted/ROI_1.nii sera créé 
ROI = util.get_ROI(brain,ROI_mask,o_name='ROI_1',o_dir='extracted')
```

```python
#Affichage de la ROI extraite
tranche = 65
plt.figure(figsize=(10,10))
plt.imshow(ROI[:,:,tranche],cmap='viridis')
plt.axis('off')
plt.title("ROI extraite, z=" + str(tranche))
plt.show()
```
